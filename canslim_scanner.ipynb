# CAN SLIM Stock Scanner

Based on William O'Neil's *How to Make Money in Stocks*

| Letter | Criteria | Threshold |
|--------|----------|-----------|
| C | Current quarterly EPS growth YoY | >= 25% |
| A | Annual EPS growth (3-yr avg) | >= 25% |
| N | Near 52-week high | Within 10% |
| S | Up/Down volume ratio | >= 1.0 |
| L | Relative Strength vs S&P 500 | RS >= 70 |
| I | Industry / Institutional ownership | Informational |
| M | Market direction (SPY above 50-day MA) | Must pass |

**How to use:** Run cells top to bottom. Edit Cell 2 to change your watchlist.

```python
# Cell 1 - Install dependencies
import subprocess, sys
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q',
                'yfinance', 'pandas', 'numpy', 'matplotlib'])
print('Dependencies ready')
```

```python
# Cell 2 - Load scanner module
# Option A: Upload canslim_scanner.py via the Files panel on the left
# Option B: Pull from GitHub (uncomment and replace YOUR_USERNAME)
# !wget -q https://raw.githubusercontent.com/YOUR_USERNAME/canslim_scanner/main/canslim_scanner.py

from canslim_scanner import CANSLIMScanner, WATCHLISTS
print('CANSLIMScanner loaded')
print('Available watchlists:', list(WATCHLISTS.keys()))
```

```python
# Cell 3 - Configure your watchlist and thresholds

MY_WATCHLIST = [
    'NVDA', 'AAPL', 'MSFT', 'GOOGL', 'META',
    'JPM',  'GS',   'V',    'MA',
    'UNH',  'LLY',
    'CAT',  'DE',
]

# Optional threshold overrides
MY_CONFIG = {
    # 'C_min_eps_growth': 20.0,
    # 'L_min_rs_rating':  60.0,
    # 'N_near_high_pct':  15.0,
}

MIN_PASSES = 4  # out of 7

print(f'Watchlist: {len(MY_WATCHLIST)} symbols')
print(f'Min passes: {MIN_PASSES}/7')
```

```python
# Cell 4 - Run the scan
scanner = CANSLIMScanner(config=MY_CONFIG)
results = scanner.scan(MY_WATCHLIST, min_passes=MIN_PASSES)
scanner.print_results(results)
```

```python
# Cell 5 - View results as DataFrame
df = scanner.to_dataframe(results)

summary = ['Symbol','Passes',
           'C_pass','A_pass','N_pass','S_pass','L_pass','I_pass','M_pass',
           'L_value','S_value','N_value']
df[summary]
```

```python
# Cell 6 - Heatmap of pass/fail results
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

criteria = ['C','A','N','S','L','I','M']
symbols  = df['Symbol'].tolist()

matrix = np.array([
    [int(df.loc[df['Symbol']==sym, f'{c}_pass'].values[0]) for c in criteria]
    for sym in symbols
])

fig, ax = plt.subplots(figsize=(10, max(4, len(symbols)*0.55)))
cmap = plt.colormaps['RdYlGn'].resampled(2)
ax.imshow(matrix, cmap=cmap, aspect='auto', vmin=0, vmax=1)
ax.set_xticks(range(len(criteria))); ax.set_xticklabels(criteria, fontsize=13, fontweight='bold')
ax.set_yticks(range(len(symbols)));  ax.set_yticklabels(symbols, fontsize=11)
for i in range(len(symbols)):
    for j in range(len(criteria)):
        ax.text(j, i, chr(10003) if matrix[i,j] else chr(10007),
                ha='center', va='center', fontsize=14)
ax.set_title('CAN SLIM Scan Results', fontsize=15, fontweight='bold')
patches = [mpatches.Patch(color=cmap(1), label='Pass'),
           mpatches.Patch(color=cmap(0), label='Fail')]
ax.legend(handles=patches, loc='upper right')
plt.tight_layout()
plt.savefig('canslim_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Cell 7 - RS Rating bar chart
import matplotlib.pyplot as plt

rs_df = df[['Symbol','L_value']].dropna().sort_values('L_value')
colors = ['#2ecc71' if v >= 70 else '#e74c3c' for v in rs_df['L_value']]

fig, ax = plt.subplots(figsize=(9, max(4, len(rs_df)*0.5)))
bars = ax.barh(rs_df['Symbol'], rs_df['L_value'], color=colors)
ax.axvline(x=70, color='orange', linestyle='--', label='RS 70 threshold')
ax.axvline(x=50, color='gray',   linestyle=':',  label='Market neutral')
for bar, val in zip(bars, rs_df['L_value']):
    ax.text(bar.get_width()+0.5, bar.get_y()+bar.get_height()/2,
            f'{val:.0f}', va='center', fontsize=10)
ax.set_xlim(0, 105)
ax.set_xlabel('RS Rating (0-99)')
ax.set_title('Relative Strength Ratings', fontsize=14, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('canslim_rs_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Cell 8 - Price chart for top pick
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

top = df.iloc[0]['Symbol'] if not df.empty else 'AAPL'
print(f'Charting: {top}')
hist = yf.Ticker(top).history(period='1y')
hist['MA50']  = hist['Close'].rolling(50).mean()
hist['MA200'] = hist['Close'].rolling(200).mean()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12,7),
                                gridspec_kw={'height_ratios':[3,1]}, sharex=True)
ax1.plot(hist.index, hist['Close'], color='#2c3e50', lw=1.5, label='Close')
ax1.plot(hist.index, hist['MA50'],  color='#3498db', lw=1.2, ls='--', label='50-day MA')
ax1.plot(hist.index, hist['MA200'], color='#e74c3c', lw=1.2, ls='--', label='200-day MA')
ax1.set_title(f'{top} - Price + Moving Averages', fontsize=14, fontweight='bold')
ax1.set_ylabel('Price ($)'); ax1.legend(); ax1.grid(alpha=0.3)
vol_colors = ['#2ecc71' if c>=o else '#e74c3c'
              for c,o in zip(hist['Close'], hist['Open'])]
ax2.bar(hist.index, hist['Volume']/1e6, color=vol_colors, alpha=0.7)
ax2.set_ylabel('Vol (M)'); ax2.grid(alpha=0.3)
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig(f'canslim_{top}_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Cell 1 - Install dependencies
import subprocess, sys
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q',
                'yfinance', 'pandas', 'numpy', 'matplotlib'])
print('Dependencies ready')
```

```python
# Cell 2 - Load scanner from GitHub
!wget -q https://raw.githubusercontent.com/kevin-mumaw/canslim_scanner/main/canslim_scanner.py

from canslim_scanner import CANSLIMScanner, WATCHLISTS
print('CANSLIMScanner loaded')
print('Available watchlists:', list(WATCHLISTS.keys()))
```

```python
# Cell 3 - Configure your watchlist
MY_WATCHLIST = [
    'NVDA', 'AAPL', 'MSFT', 'GOOGL', 'META',
    'JPM',  'GS',   'V',    'MA',
    'UNH',  'LLY',
    'CAT',  'DE',
]

MY_CONFIG = {}
MIN_PASSES = 4  # out of 7

print(f'Watchlist: {len(MY_WATCHLIST)} symbols')
print(f'Min passes: {MIN_PASSES}/7')
```

```python
# Cell 4 - Run the scan
scanner = CANSLIMScanner(config=MY_CONFIG)
results = scanner.scan(MY_WATCHLIST, min_passes=MIN_PASSES)
scanner.print_results(results)
```
